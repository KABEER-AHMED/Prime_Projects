```python
"""
q1 create a banaccount class
with attributes account_number and owner_number and balance
and then add methods to deposit, withdraw and check balance
"""

class BankAccount:
    def __init__(self,account_number,owner_name,balance ):
        self.account_number = account_number
        self.owner_name = owner_name
        self.balance = balance
        
    
    def deposit(self,money):
        self.balance += money
    
    def withdraw(self,amount):
        if (self.balance - amount) < 0 :
            print("not enough balance fair ke bacche !!!")
        else:
            self.balance -= amount
    
    def check_balance(self):
        print(f"the remaining balance is {self.balance}")
    

person1 = BankAccount(0,"kabeer",1000)
person1.deposit(1500)
person1.withdraw(1000)
person1.check_balance()

```

```python
""" q2 crete a class book with the following attributes

title, author, list of reviews, 
and add methods to 
add a new review, count reviews, display all reviews
"""

reviews = ["great product", "fantastic product", "couldve been made better"]
class Book:
    def __init__(self, title, author, reviews):
        self.title = title
        self.author = author
        self.reviews = reviews
    
    def add_review(self,review):
        self.reviews.append(review)
    def review_count(self):
        i = 0
        for x in self.reviews:
            i+=1
        print(f"the total number of reviews are {i}")
    def review_display(self):
        print(" these are all the reviews : ")
        for i in self.reviews:
            print(i)
        
    

book1 = Book("andmand kka tona shona", "kabeer", reviews)
book1.add_review("bakkkwas book hai")
book1.review_count()
book1.review_display()
```

```python
"""q9 create the following classes 
Herbivore , Carnivore, Omnivore wwith some attributes and methods
then create a class Bear that inherits from all the above classes to
showcase how multiple inheritance works"""

class Herbivore:
    def __init__(self,name, *args,**kwargs):
        self.name = name
        super().__init__(*args,**kwargs)
    def eat(self):
        print("the herbivore is eats plants")

class Carnivore:
    def __init__(self,name,*args,**kwargs):
        self.name = name
        super().__init__(*args,**kwargs)
    def eat(self):
        print("the carnivore eats meat")

class Omnivore:
    def __init__(self,name,*args,**kwargs):
        self.name = name
        super().__init__(*args,**kwargs)
    def eat(self):
        print("the omnivore eats both plant and meat")

class Bear(Herbivore,Carnivore,Omnivore):
    def __init__(self,name1,name2,name3,name,*args,**kwargs):
        self.name = name
        super().__init__(name1,name2,name3,*args,**kwargs)
    def bear_attack(self):
        print("bear attacks")
        
        
        


h = Herbivore("herbi")
c = Carnivore("carni")
o = Omnivore("omni")
b = Bear("hara","cara","ora","bhalu")

print(h.name,c.name,o.name,b.name)
```
