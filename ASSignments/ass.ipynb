```python
"""
q1 create a banaccount class
with attributes account_number and owner_number and balance
and then add methods to deposit, withdraw and check balance
"""

class BankAccount:
    def __init__(self,account_number,owner_name,balance ):
        self.account_number = account_number
        self.owner_name = owner_name
        self.balance = balance
        
    
    def deposit(self,money):
        self.balance += money
    
    def withdraw(self,amount):
        if (self.balance - amount) < 0 :
            print("not enough balance fair ke bacche !!!")
        else:
            self.balance -= amount
    
    def check_balance(self):
        print(f"the remaining balance is {self.balance}")
    

person1 = BankAccount(0,"kabeer",1000)
person1.deposit(1500)
person1.withdraw(1000)
person1.check_balance()

```

```python
"""q9 create the following classes 
Herbivore , Carnivore, Omnivore wwith some attributes and methods
then create a class Bear that inherits from all the above classes to
showcase how multiple inheritance works"""

class Herbivore:
    def __init__(self,name, *args,**kwargs):
        self.name = name
        super().__init__(*args,**kwargs)
    def eat(self):
        print("the herbivore is eats plants")

class Carnivore:
    def __init__(self,name,*args,**kwargs):
        self.name = name
        super().__init__(*args,**kwargs)
    def eat(self):
        print("the carnivore eats meat")

class Omnivore:
    def __init__(self,name,*args,**kwargs):
        self.name = name
        super().__init__(*args,**kwargs)
    def eat(self):
        print("the omnivore eats both plant and meat")

class Bear(Herbivore,Carnivore,Omnivore):
    def __init__(self,name1,name2,name3,name,*args,**kwargs):
        self.name = name
        super().__init__(name1,name2,name3,*args,**kwargs)
    def bear_attack(self):
        print("bear attacks")
        
        
        


h = Herbivore("herbi")
c = Carnivore("carni")
o = Omnivore("omni")
b = Bear("hara","cara","ora","bhalu")

print(h.name,c.name,o.name,b.name)
```

# Working with data

- 1. data is inconsistent ------- clean it first
- 2. missing data   ------------- fill them up
- 3. duplicate data ------------- handle duplicates

```python
def load_data(filename):
    with open(filename) as f:
        data = json.load(f)
    return data
    

data = load_data("amazon_store_data.json")

print(data)
```

```python
# cleaning the data 
def clean_data(pr):
    cleans_data = []
    unique=set()
    text_to_num = {"one" : 1,"two":2,"three":3,"four":4,"five":5}
    for user in pr:
        #cleaning the ratings
        raw_rating = user["rating"].strip().lower()
        if raw_rating in text_to_num:
            user["rating"] = text_to_num[raw_rating]

        # handling missing values
        raw_age = user.get("age") #since get function return none is nothing
        if raw_age == None:
            user["age"] = None
        
        # de duplicaion
        if user["name"] in unique:
            continue
        else:
            unique.add(user["name"])
        
        cleans_data.append(user)
        return cleans_data

x= clean_data(data)
print(x)
```
